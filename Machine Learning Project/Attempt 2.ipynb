## Balance the cohorts using SMOTE

```
from imblearn.over_sampling import SMOTE
oversample = SMOTE()
variables, cohort = oversample.fit_resample(df_noid.loc[:, df_noid.columns != "label"], df_noid["label"])
cohort.value_counts("label")
```

Cohorts now balanced

```
df = variables.append(cohort)
```

## Determine the relationship between columns and cohort

```
percent = normalized_df.groupby("label").mean().pct_change() * 100
percent = percent.abs()
percent
```

Choosing the columns with a more than 50% difference. 

```
chosen_variables = percent.loc[:, (percent > 50).any()]
list(chosen_variables.columns)
```

```
variables = normalized_df[[
 'skill6',
 'play_measure6',
 'location3',
 'social1',
 'location2',
 'skill7',
 'play_measure4',
 'play_measure2',
 'skill4',
 'skill12',
 'skill5',
 'location4',
 'skill13']]
variables.head() # X
```

```
cohort = normalized_df["label"]
cohort.shape # Y
```

```
cohort.value_counts("label")
```

```
from sklearn.model_selection import train_test_split
cohort_train, cohort_test, variables_train, variables_test = train_test_split(cohort,variables,train_size=0.3, stratify=cohort)
```

```
cohort_test.value_counts("label")
```

```
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=4000)
```

```
model.fit(variables_train, cohort_train)
```

```
model.predict(variables_test)
```

```
model.score(variables_test,cohort_test)
```
