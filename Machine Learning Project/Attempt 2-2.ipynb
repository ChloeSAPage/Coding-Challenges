# Data Preparation

## Load data frame

```
import pandas as pd
df = pd.read_csv("modelling_challenge.csv")
df.columns
```

```
df.info()
```

do they need to all be floats?

```
df.describe()
```

## NaN values

```
df.isna().sum()
```

```
df.shape
```

```
df_no_nan = df.dropna() # Opting for removal of NaN values
df_no_nan.shape
```

```
# min-max normalisation
normalized_df=(df_no_nan-df_no_nan.min())/(df_no_nan.max()-df_no_nan.min())
normalized_df
```

```
df_noid =  df_no_nan.drop("player_id", axis=1)
df_noid
```

## Balance the cohorts using SMOTE

```
from imblearn.over_sampling import SMOTE
oversample = SMOTE()
variables, cohort = oversample.fit_resample(df_noid.loc[:, df_noid.columns != "label"], df_noid["label"])
cohort.value_counts("label")
```

Cohorts now balanced

```
variables.shape
```

```
cohort.shape
```

```
df = variables.join(cohort)
df.value_counts("label")
```

## Determine the relationship between columns and cohort

```
percent = df.groupby("label").mean().pct_change() * 100
percent = percent.abs()
percent
```

Choosing the columns with a more than 50% difference. 

```
chosen_variables = percent.loc[:, (percent > 50).any()]
chosen_variables = list(chosen_variables.columns)
chosen_variables
```

```
variables = df[chosen_variables]
variables.shape # X
```

```
# cohort = normalized_df["label"]
cohort = df["label"]
cohort.shape # Y
```

```
cohort.value_counts("label")
```

# Models

## Split the data into training and test set

```
from sklearn.model_selection import train_test_split
cohort_train, cohort_test, variables_train, variables_test = train_test_split(cohort,variables,train_size=0.75, stratify=cohort)
```

```
cohort_test.value_counts("label") # Still balanced
```

## K-Nearest Neighbour

```
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(variables_train, cohort_train)
model.predict(variables_test)
model.score(variables_test,cohort_test)
```

## Random Forest

```
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(variables_train, cohort_train)
model.predict(variables_test)
model.score(variables_test,cohort_test)
```

## Decision Tree

```
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(variables_train, cohort_train)
model.predict(variables_test)
model.score(variables_test,cohort_test)
```
